```python
#Exo 2.11
nb1 = int(input("Entrez un nombre: "))
 
n1 = 0
n2 = 1
 
print("\n La suite fibonacci est :")
print(n1, ",", n2, end=", ")
 
for i in range(2, nb1):
  n3 = n1 + n2
  print(n3, end=", ")
 
  n1 = n2
  n2 = n3
```
